# Creating New Columns For Analysis

# Creating Sales Category

Sales categories are created to classify stores based
on weekly sales performance.

```python
# Creating sales category

df['Sales_Category'] = pd.cut(df['Weekly_Sales'],
                              bins=[0, 5000, 20000, 80000],
                              labels=['Low', 'Medium', 'High'])

df[['Weekly_Sales', 'Sales_Category']].head()
```

# Creating Season Column

Season categories are created based on months
to analyze seasonal sales trends.

```python
# Creating season column

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8]:
        return 'Monsoon'
    else:
        return 'Autumn'

df['Season'] = df['Month'].apply(get_season)

df[['Month', 'Season']].head()
```

# Creating Holiday Labels

Holiday labels are created for better understanding
during visualization.

```python
# Creating holiday labels

df['Holiday_Label'] = df['IsHoliday'].map({0:'Non-Holiday',
                                           1:'Holiday'})

df[['IsHoliday', 'Holiday_Label']].head()
```

# Exploratory Data Analysis and Visualizations

Exploratory Data Analysis (EDA) is performed to identify
sales trends, customer behavior, store performance,
and relationships between different variables using
data visualization techniques.

## Weekly Sales Distribution Analysis

This visualization is used to understand the distribution and frequency of weekly sales values across the dataset. It helps identify sales concentration and extreme sales values.

```python
# Distribution plot of weekly sales

plt.figure(figsize=(10,5))

sns.histplot(df['Weekly_Sales'],
             bins=50,
             kde=True)

plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')

plt.show()
```

### Observation

The distribution shows that most weekly sales values are concentrated in the lower range, while a few records have very high sales values.

## Store-wise Sales Visualization

This visualization compares the total sales generated by different stores to identify high-performing and low-performing stores.

```python
# Store-wise total sales visualization

store_sales = df.groupby('Store')['Weekly_Sales'].sum().sort_values()

plt.figure(figsize=(14,6))

store_sales.plot(kind='bar')

plt.title('Store-wise Total Sales')
plt.xlabel('Store')
plt.ylabel('Total Sales')

plt.show()
```

### Observation

Some stores generate significantly higher sales compared to others, indicating differences in store performance and customer demand.

## Department-wise Sales Visualization

This visualization identifies departments contributing the highest sales revenue within the business.

```python
# Department-wise sales visualization

dept_sales = df.groupby('Dept')['Weekly_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))

sns.barplot(x=dept_sales.index,
            y=dept_sales.values)

plt.title('Top 10 Departments by Sales')
plt.xlabel('Department')
plt.ylabel('Total Sales')

plt.show()
```

### Observation

Certain departments contribute more to overall revenue, indicating high customer demand for specific product categories.

## Monthly Sales Trend Analysis

This visualization shows how sales vary across different months and helps identify seasonal sales trends.

```python
# Monthly sales trend

monthly_sales = df.groupby('Month')['Weekly_Sales'].sum()

plt.figure(figsize=(10,5))

sns.lineplot(x=monthly_sales.index,
             y=monthly_sales.values,
             marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.show()
```

### Observation

Sales fluctuate across different months, showing seasonal purchasing patterns and changing customer demand.

## Holiday Sales Comparison

This visualization compares average sales during holiday and non-holiday periods to understand holiday sales impact.

```python
# Holiday vs Non-Holiday sales

holiday_sales = df.groupby('Holiday_Label')['Weekly_Sales'].mean()

plt.figure(figsize=(6,5))

sns.barplot(x=holiday_sales.index,
            y=holiday_sales.values)

plt.title('Holiday vs Non-Holiday Sales')
plt.xlabel('Holiday Type')
plt.ylabel('Average Weekly Sales')

plt.show()
```

### Observation

The comparison helps understand how holiday periods impact customer purchasing behavior and sales performance.

## Store Type Sales Analysis

This visualization compares the sales performance of different store categories such as Type A, B, and C.

```python
# Sales by store type

type_sales = df.groupby('Type')['Weekly_Sales'].sum()

plt.figure(figsize=(6,5))

sns.barplot(x=type_sales.index,
            y=type_sales.values)

plt.title('Sales by Store Type')
plt.xlabel('Store Type')
plt.ylabel('Total Sales')

plt.show()
```

### Observation

Different store categories contribute differently to total sales revenue.

## Correlation Heatmap Analysis

## Seasonal Sales Analysis
